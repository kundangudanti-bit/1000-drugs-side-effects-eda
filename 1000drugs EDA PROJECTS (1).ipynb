# univariate analysis

```python
bins = [18, 25, 35, 45, 55, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55+']

data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

age_group_counts = data['Age_Group'].value_counts().sort_index()

plt.bar(age_group_counts.index, age_group_counts.values)
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Group Distribution")
plt.show()
```

```python

sns.histplot(
    data['Improvement_Score'],
    bins=15,
    kde=True,
    color='teal'
)
plt.xlabel("Improvement Score")
plt.ylabel("Count")
plt.title("Improvement Score Distribution")
plt.show()

```

```python
data['Side_Effects'].value_counts().plot(kind='bar')
plt.title("Frequency of Side Effects")
plt.xlabel("Side Effects")
plt.ylabel("Count")
plt.show()
```

# bi varient analysis

#  numerical vs category

```python
drug_mean = data.groupby('Drug_Name')['Improvement_Score'].mean()

plt.bar(drug_mean.index, drug_mean.values)
plt.xlabel("Drug Name")
plt.ylabel("Average Improvement Score")
plt.title("Average Improvement Score by Drug")
plt.xticks(rotation=90)
plt.show()
```

```python
data.boxplot(column='Improvement_Score', by='Condition')
plt.title("Improvement Score by Condition")
plt.suptitle("")
plt.xlabel("Condition")
plt.ylabel("Improvement Score")
plt.show()
```

# category vs category

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.countplot(
    data=data,
    x='Condition',
    hue='Side_Effects',
    palette='tab20'
)
plt.xlabel("Condition")
plt.ylabel("Count")
plt.title("Side Effects Distribution by Condition")
plt.legend(bbox_to_anchor=(1.05, 1), title="Side Effects")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    data=data,
    x='Side_Effects',
    hue='Gender'
)
plt.xlabel("Side Effects")
plt.ylabel("Count")
plt.title("Side Effects by Gender")
plt.xticks(rotation=90)
plt.legend(title="Gender")
plt.show()
```

# numerical vs numerical

```python
plt.scatter(data['Age'], data['Improvement_Score'])
plt.xlabel("Age")
plt.ylabel("Improvement Score")
plt.title("Age vs Improvement Score")
plt.show()
```

```python
avg_dosage = data.groupby('Dosage_mg')['Improvement_Score'].mean()

plt.plot(avg_dosage.index, avg_dosage.values)
plt.xlabel("Dosage (mg)")
plt.ylabel("Average Improvement Score")
plt.title("Average Improvement Score by Dosage")
plt.show()



```

# multivariate analysis

```python
# Select numerical columns
num_data = data[['Age', 'Dosage_mg', 'Treatment_Duration_days', 'Improvement_Score']]

# Compute correlation matrix
corr = num_data.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
